```python
import pandas as pd
import numpy as np
import os
from bs4 import BeautifulSoup
from selenium import webdriver
import sys
import json
import csv
import re
from time import sleep
```

```python
chromedriver = "C:/Users/HP/Downloads/chromedriver_win32/chromedriver" # path to the chromedriver executable
chromedriver = os.path.expanduser(chromedriver)
print('chromedriver path: {}'.format(chromedriver))
sys.path.append(chromedriver)
driver = webdriver.Chrome(chromedriver)
```

```python
hrefs=[]
head = ['URL']
def get_page(soup):
#     if soup.find("span", {"class": "css-1uq0cfn"}) is not None: 
#         print("Found")
#         for link in soup.find("span", {"class": "css-1uq0cfn"}).findAll("a"):
#             print(link.get("href"))
#             hrefs.append(str(link.get("href")))
    for a in soup.find_all('a', href=True): 
        if a.text:
            temp1 = a['href'].split('/')
            temp2 = a['href'].split('?')
            if(len(temp2)==1 and len(temp1)>1 and temp1[1]=='biz'):
                url_not_recommended = a['href'].replace("biz", "not_recommended_reviews")
                hrefs.append("https://m.yelp.com"+url_not_recommended)
    else:
        print("page over")

def hrefcsv(hrefs):    
    with open("apartmentURL_FAKE.csv",'r+') as f:
        writer = csv.writer(f)
        writer.writerow(head)
        for i in hrefs:
            writer.writerow([i])
        f.close()
    
        
```

```python
pages = [0,10,20]
for i in pages:
    driver.get("https://www.yelp.com/search?cflt=apartments&find_loc=San+Francisco,+CA&start="+str(i))
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    get_page(soup)

hrefcsv(hrefs)
```

```python

date = []
username = []
date = []
location = []
review_count = []
friend_count = []
photo_count =[]
rating = []
comments = []

def get_fake_comments(url): 
    
    driver = webdriver.Chrome('C:/Users/HP/Downloads/chromedriver_win32/chromedriver')
    
    # create url for "not recommended" page based on original restaurant URL 
    url_not_recommended = url.replace("biz", "not_recommended_reviews")
    
    # launch scraper on URL
    driver.get(url_not_recommended)
    sleep(10)
    
    # create list for all DF columns
    
    
    
    # while loop used in order to iterate to "next pages"
    # main comment block
    while True:
        bad_comment_blocks = driver.find_elements_by_css_selector(".not-recommended-reviews")[0].find_elements_by_css_selector(".review.review--with-sidebar")
        
        # iterate through comment block to get all comment features for 1 page


        for i in range(0, (len(bad_comment_blocks))):

            try:
                comments.append(bad_comment_blocks[i].find_element_by_tag_name("p").text)
            except:
                comments.append("N/A")


            try:
                rating.append(bad_comment_blocks[i].find_elements_by_css_selector(".i-stars")[0].get_attribute("title").split()[0])

            except: 
                rating.append("N/A")


            try:
                username.append(bad_comment_blocks[i].find_elements_by_css_selector(".user-display-name")[0].text)
            except: 
                username.append("Not Yelp User")

            try:
                review_count.append(bad_comment_blocks[i].find_elements_by_css_selector(".review-count")[0].text.split()[0])
            except: 
                review_count.append("N/A")

            try:
                friend_count.append(bad_comment_blocks[i].find_elements_by_css_selector(".friend-count")[0].text.split()[0])
            except: 
                friend_count.append("N/A")


            try: 
                location.append(bad_comment_blocks[i].find_elements_by_css_selector(".user-location")[0].text)
            except: 
                location.append("N/A")


            try: 
                date.append(bad_comment_blocks[i].find_elements_by_css_selector(".rating-qualifier")[0].text)
            except: 
                date.append("N/A")
            
            try:
                photo_count.append(bad_comment_blocks[i].find_elements_by_css_selector(".photo-count")[0].text.split()[0])
            except: 
                photo_count.append("N/A")
            
            
        try:
            #navigate to the next page
            driver.find_elements_by_css_selector(".not-recommended-reviews")[0].find_elements_by_css_selector('.icon.icon--24-chevron-right.icon--size-24')[0].click()
            print(url_not_recommended, 'subpage scraped')
            sleep(7) 
        except:
            ()
            break

    # make dict of all lists, order as per same scraper for recommended comments
    # add some columns with constant values

    
 

    


#     return data
#     print(len(data))
#     print(data.describe())
```

```python
dd = {"date" : date,
         "username" : username,
         "photos_for_review" : [-1] * len(date),
         "rating" : rating,
         "text_review" : comments, 
         "user_location" : location, 
         "user_friends_count" : friend_count, 
         "user_reviews_count" : review_count, 
         "user_total_image_posted" : photo_count, 
         "is_real_review" : [0] * len(date), 
         
        }
         
         
data = pd.DataFrame(dd)
print(len(data))
print(data.describe())
```
